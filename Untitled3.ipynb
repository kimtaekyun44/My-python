```python
import smtplib
import os
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
```

```python
SMTP_SERVER = "smtp.gmail.com"  # Gmail SMTP 서버
SMTP_PORT = 587  # TLS 포트 (465는 SSL)
GMAIL_USER = "kungom75@gmail.com"  # 본인의 이메일 주소
GMAIL_PASSWORD = "ernb zqrn hcqy qoxd"  # Gmail 앱 비밀번호
```

```python
TO_EMAIL = "kungom@naver.com"  # 받는 사람 이메일
SUBJECT = "첨부파일이 포함된 이메일 전송 테스트"
BODY = "안녕하세요,\n\n첨부파일과 함께 이메일을 전송합니다.\n\n감사합니다."
```

```python
# ATTACHMENT_PATH = r"c:\python\humanmatrix_introdutioni.pdf"
ATTACHMENT_PATH = r"c:\python\휴먼메트릭스_채용대행_소개자료.pdf"
```

```python
os.path.exists(ATTACHMENT_PATH)
```

```python
msg = MIMEMultipart()
msg["From"] = GMAIL_USER
msg["To"] = TO_EMAIL
msg["Subject"] = SUBJECT
msg.attach(MIMEText(BODY, "plain"))
```

```python
if os.path.exists(ATTACHMENT_PATH):
    with open(ATTACHMENT_PATH, "rb") as attachment:
        part = MIMEBase("application", "octet-stream")
        part.set_payload(attachment.read())
        encoders.encode_base64(part)
        part.add_header(
            "Content-Disposition",
            f'attachment; filename="{os.path.basename(ATTACHMENT_PATH)}"; filename*=UTF-8\'\'{os.path.basename(ATTACHMENT_PATH)}',
        )
        msg.attach(part)
else:
    print(f"첨부 파일을 찾을 수 없습니다: {ATTACHMENT_PATH}")

# SMTP 서버 연결 및 메일 전송
try:
    server = smtplib.SMTP(SMTP_SERVER, SMTP_PORT)
    server.starttls()  # TLS 보안 연결
    server.login(GMAIL_USER, GMAIL_PASSWORD)
    server.sendmail(GMAIL_USER, TO_EMAIL, msg.as_string())
    server.quit()
    print("이메일 전송 성공!")
except Exception as e:
    print(f"이메일 전송 실패: {e}")
```
