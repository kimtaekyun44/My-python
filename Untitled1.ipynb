```python
import requests
from bs4 import BeautifulSoup

# 웹페이지 URL
url = "https://booktoki466.com/novel/678978"

# 요청 보내기
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
}
response = requests.get(url, headers=headers)

# 응답 확인
if response.status_code == 200:
    # BeautifulSoup로 HTML 파싱
    soup = BeautifulSoup(response.text, "html.parser")
    
    # 원하는 데이터 추출 (예: 제목이나 본문)
    title = soup.find("title").text  # 제목 추출
    print(f"Title: {title}")
    
    # 본문 데이터가 있는 태그 검색
    content = soup.find("div", {"class": "content"})  # 적절한 태그/클래스 수정
    if content:
        print(content.text)
    else:
        print("본문 데이터를 찾을 수 없습니다.")
else:
    print(f"Failed to fetch the webpage. Status code: {response.status_code}")
```

```python
import requests

API_KEY = "YOUR_2CAPTCHA_API_KEY"  # 2Captcha API 키
CAPTCHA_IMAGE_URL = "CAPTCHA 이미지 URL"

# 2Captcha에 이미지 전송
response = requests.post("http://2captcha.com/in.php", data={
    "key": API_KEY,
    "method": "base64",
    "body": CAPTCHA_IMAGE_URL,
    "json": 1
})
captcha_id = response.json().get("request")

# CAPTCHA 해결 결과 가져오기
result_url = f"http://2captcha.com/res.php?key={API_KEY}&action=get&id={captcha_id}&json=1"
while True:
    result = requests.get(result_url).json()
    if result.get("status") == 1:
        captcha_solution = result.get("request")
        break

print(f"CAPTCHA 해결 완료: {captcha_solution}")
```
