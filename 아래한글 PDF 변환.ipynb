```python
import os
import win32com.client as win32

```

```python
# 한글 프로그램 실행
hwp = win32.gencache.EnsureDispatch("HWPFrame.HwpObject")
hwp.RegisterModule("FilePathCheckDLL", "FilePathCheckerModule")

# 변환할 파일 경로 설정
getPath = r'C:\Users\user\OneDrive\바탕 화면\download'
files = [f for f in os.listdir(getPath) if f.endswith(('.hwp', '.hwpx'))]

```

```python
# 일괄 변환 실행
for file in files:
    hwp.Open(os.path.join(getPath, file))
    pre, ext = os.path.splitext(file)
    
    # PDF 저장 설정
    hwp.HAction.GetDefault("FileSaveAs_S", hwp.HParameterSet.HFileOpenSave.HSet)
    hwp.HParameterSet.HFileOpenSave.filename = os.path.join(getPath, f"{pre}.pdf")
    hwp.HParameterSet.HFileOpenSave.Format = "PDF"
    hwp.HAction.Execute("FileSaveAs_S", hwp.HParameterSet.HFileOpenSave.HSet)

hwp.Quit()
```
