```python
import tkinter as tk
from tkinter import filedialog
import fitz  # PyMuPDF
from IPython.display import display, HTML
import base64
import io
from PIL import Image

def display_pdf():
    # tkinter root 윈도우 생성 및 숨기기
    root = tk.Tk()
    root.withdraw()
    
    # 파일 선택 다이얼로그 열기
    file_path = filedialog.askopenfilename(
        title="PDF 파일을 선택하세요",
        filetypes=[("PDF files", "*.pdf")]
    )
    
    if not file_path:
        print("파일이 선택되지 않았습니다.")
        return
    
    try:
        # PDF 문서 열기
        pdf_document = fitz.open(file_path)
        
        # 각 페이지를 이미지로 변환하여 표시
        for page_num in range(min(pdf_document.page_count, 5)):  # 처음 5페이지만 표시
            page = pdf_document[page_num]
            
            # 페이지를 이미지로 변환 (해상도 향상)
            pix = page.get_pixmap(matrix=fitz.Matrix(2, 2))
            
            # PIL 이미지로 변환
            img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
            
            # 이미지를 바이트로 변환
            img_byte_arr = io.BytesIO()
            img.save(img_byte_arr, format='PNG')
            img_byte_arr = img_byte_arr.getvalue()
            
            # Base64로 인코딩
            img_base64 = base64.b64encode(img_byte_arr).decode()
            
            # HTML로 표시
            display(HTML(f"""
                <div style="margin-bottom: 20px;">
                    <h3>Page {page_num + 1}</h3>
                    <img src="data:image/png;base64,{img_base64}" style="border: 1px solid #ddd; max-width: 100%;">
                </div>
            """))
        
        if pdf_document.page_count > 5:
            print(f"\n처음 5페이지만 표시되었습니다. 전체 페이지 수: {pdf_document.page_count}")
        
        pdf_document.close()
        
    except Exception as e:
        print(f"에러가 발생했습니다: {str(e)}")
    
    finally:
        # tkinter root 윈도우 제거
        root.destroy()

# Jupyter Notebook에서 실행할 셀
# display_pdf()
```

```python
display_pdf()
```
