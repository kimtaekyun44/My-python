```python
import pandas as pd
```

```python
df = pd.read_excel('메타버스 박물관 관련 인식조사 및 이용의도 설문_Rawdata_20241118_데이터.xlsx')
```

```python
df['q1'].value_counts().sort_index().to_frame()
```

```python
def makeFrame(category):
    df2 = df[category].value_counts().sort_index().to_frame()
    print(df2)
    df2 = df2.reset_index()
    df2.columns = ['name' , 'count']
    # 전체 합계를 계산
    total_count = df2['count'].sum()
    df2['p_count'] = df2['count'] / total_count * 100
    df2['p_count'] = df2['p_count'].astype('str') + '%'
    df2['category'] = category
    return df2
```

```python
df3 = pd.DataFrame(columns = ['category','name','count','p_count'])
categorys = df.columns[1:]
for category in categorys:
    df2 = makeFrame(category)
    df3 = pd.concat([df3,df2])
```

```python
df3.to_csv('bbb.csv',encoding='euc-kr')
```
