```python
import cv2
import os

def capture_frames(video_path, output_folder, interval=5):
    # 비디오 파일 열기
    cap = cv2.VideoCapture(video_path)

    # 비디오의 FPS(Frame Per Second)와 전체 프레임 수 가져오기
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # 5초 단위로 캡처할 프레임 계산
    frame_interval = int(fps * interval)  # interval 초에 해당하는 프레임 수

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    frame_count = 0
    captured_count = 0

    while True:
        ret, frame = cap.read()

        # 비디오의 끝에 도달했을 때 종료
        if not ret:
            break

        # 지정된 간격에 해당하는 프레임을 저장
        if frame_count % frame_interval == 0:
            output_path = os.path.join(output_folder, f"frame_{captured_count:04d}.jpg")
            cv2.imwrite(output_path, frame)
            print(f"Captured frame at {frame_count/fps:.2f} seconds: {output_path}")
            captured_count += 1

        frame_count += 1

    cap.release()
    print("Frame capture completed.")

```

```python
# 예제 실행
capture_frames("아파트아파트.mp4", "./captured_frames", interval=5)
```
